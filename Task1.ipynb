# **Importing pandas, reading the tab-separated file 'marketing_campaign.csv' into a DataFrame called df, and then displaying the first few rows using df.head() to quickly check the data.**

```python
import pandas as pd

df = pd.read_csv('marketing_campaign.csv', sep='\t')
df.head()

```

# **Checking for null values and counting them in each column.**

```python
df.isnull().sum()
```

# **Filling null values as Unknown.**

```python
df.fillna({'Education': 'Unknown', 'Marital_Status': 'Unknown'}, inplace=True)
```

# **Removing Null values from the dataset**

```python
df.drop_duplicates(inplace=True)
```

# **Standardizing the text values**

```python
df['Education'] = df['Education'].replace({
    'Graduation': 'Graduate',
    'PhD': 'Doctorate',
    'Master': 'Postgraduate',
    '2n Cycle': 'Postgraduate',
    'Basic': 'School'
})
```

```python
df['Marital_Status'] = df['Marital_Status'].replace({
    'Married': 'Married',
    'Together': 'Married',
    'Single': 'Single',
    'Divorced': 'Separated',
    'Widow': 'Separated',
    'Alone': 'Single',
    'Absurd': 'Single',
    'YOLO': 'Single'
})
```

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y', errors='coerce')
```

```python
df.head()
```

# **Coverting date Format**

```python
df['Dt_Customer'] = df['Dt_Customer'].dt.strftime('%d-%m-%Y')
```

```python
df.head()
```

# **Renaming column Headers**

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('-', '_')
```

```python
df.head()
```

# **Checking the datatypes**

```python
df.dtypes
```
